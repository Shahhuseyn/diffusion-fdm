## Task 1.1

According to **Fick's second law**, the diffusion flux is proportional to the negative gradient of concentration:

$$ \boldsymbol{j}(x',y',t) = -\boldsymbol{D'} \cdot \boldsymbol{\nabla} \boldsymbol{\rho}(x',y',t)$$

Where $\boldsymbol{D'}$ is the symmetrical diffusion tensor of proportionality constants for a two-dimensional system with anisotropic diffusion properties where the diffusion is defined in the generalized coordinates $(x', y')$:

$$\boldsymbol{D'} = \begin{bmatrix}
D'_{xx} & D'_{xy} \\
D'_{yx} & D'_{yy}\end{bmatrix} $$


$D'$ is an symmetrical tensor $\rightarrow$ $D'_{xy}=D'_{yx}$ and also to simplify the calculations it will be assumed that diffusion occurs independently along the principal axes $(x,y)$ of the material. This implies a coordinate transformation that diagonalizes the diffusion tensor, resulting in:

$$\boldsymbol{D} = \begin{bmatrix} D_{xx} & 0 \\0 & D_{yy}\end{bmatrix} $$
Where $ D_{xx} = \frac{(D'_{xx} + D'_{yy}) + \sqrt{(D'_{xx}-D'_{yy})^2 + 4 \cdot (D'_{xy})^2} }{2} $ and $ D_{yy} = \frac{(D'_{xx} + D'_{yy}) - \sqrt{(D'_{xx}-D'_{yy})^2 + 4 \cdot (D'_{xy})^2} }{2} $.



Next, we substitute this diagonalized tensor into Fick’s first law to determine the flux $\boldsymbol{j}(x,y)$ in the principal coordinate system:

$$ \boldsymbol{j}(x,y,t) = -\boldsymbol{D} \cdot \boldsymbol{\nabla} \boldsymbol{\rho}(x,y,t)$$

$$ \boldsymbol{j}(x,y,t) = - \begin{bmatrix}D_{xx} & 0 \\0 & D_{yy}\end{bmatrix} \cdot \begin{bmatrix}\frac{\partial}{\partial x} \\ \frac{\partial}{\partial y}\end{bmatrix} \boldsymbol{\rho}(x,y,t)$$


$$\begin{bmatrix}j_x \\j_y\end{bmatrix} = \begin{bmatrix}-D_{xx}\frac{\partial\boldsymbol{\rho}(x,y,t)}{\partial x} \\ - D_{yy}\frac{\partial\boldsymbol{\rho}(x,y,t)}{\partial y}\end{bmatrix} $$

<br><br>

The **continuity equation** takes the following form for the case where the sink and source terms are neglected:
$$ \frac{\partial\boldsymbol{\rho}(x,y,t)}{\partial t} + \boldsymbol{\nabla} \cdot \boldsymbol{j}(x,y,t) = 0$$

<br><br>

Substituing the $\boldsymbol{j}(x,y,t)$ to the contuinuty equation and rearranging the terms, the equation becomes:

$$ \frac{\partial \rho}{\partial t} = \frac{\partial}{\partial x} \left( D_{xx} \frac{\partial \rho}{\partial x} \right) + \frac{\partial}{\partial y} \left( D_{yy} \frac{\partial \rho}{\partial y} \right) $$ 

<br><br>

$$\frac{\partial \rho}{\partial t} = D_{xx} \frac{\partial^2 \rho}{\partial x^2} + D_{yy} \frac{\partial^2 \rho}{\partial y^2} \tag{1}$$

Which is the Ficks second law in 2-D under the assumption that the material has anisotropic diffusion properties where source or sink terms are neglected.

<br><br>

Classification of PDEs:

The partial differential equation exhibits first-order temporal derivative ( the equation includes a first-order derivative with respect to time $\frac{\partial \rho}{\partial t}$) indicating a temporal evolution of the solution, second-order spatial derivatives (it includes second-order derivatives with respect to spatial variables -  $\frac{\partial^2 \rho}{\partial x^2}$, $\frac{\partial^2 \rho}{\partial y^2}$) which describe diffusive effects. Furthermore there is no second-order temporal derivative which differentiates it from hyperbolic equations. These characteristics define this PDE as **Parabolic** as it models diffusion - a processe where the rate of change in time is proportional to the second-order spatial variations. Furthermore, the equation closely resembles the classical heat equation:

$$ \frac{\partial \Theta}{\partial t} = D \cdot \boldsymbol \nabla^2 \Theta$$

which is a canonical example of a parabolic PDE, further supporting its classification.



## Task 1.2 
For the stationary (steady-state) diffusion case, the temporal derivative in equation (1) becomes zero $ \left( \frac {\partial \rho}{\partial t} = 0 \right)$. If both sides of the equation are divided by **$D_{xx}$**, and $\frac {D_{yy}}{D_{xx}}$ is defined as $P$ the equation simplifies to the following form:

<br><br>

$$ \partial^2_{xx} \rho (x,y) + P \partial^2_{yy} \rho(x,y)=0 \tag{2}$$

<br><br>

x and y indicates the two spatial coordinates in the plane along which diffusion takes place, and they are aligned with the principal axes of the material’s anisotropy.
<br><br>
The parameter $P$, which can be referred as a dimensionless anisotropy ratio or principal Diffusivity ratio, encodes how much faster (or slower) diffusion is in the y‐direction compared to the x‐direction.
* The general anisotropic diffusion tensor $D'$ is diagonalized (in 2D, by rotating into its principal directions). In that diagonal form, the ratio of the diagonal entries $\frac{D_{yy}}{D_{xx}}$ becomes $P$.
* If ${D_{yy}} = {D_{xx}}$ then $P = 1$, and the stationary equation reduces to the Isotropic Laplace equation $\rightarrow \nabla^2 \rho = 0$.
* If ${D_{yy}} \neq {D_{xx}}$, diffusion is faster along one principal axis than the other, and $P \neq 1$.

## Task 1.3
In order to transform the equation (2) to the standard Laplace equation with new variables of $\widetilde{x}$ and $\widetilde{y}$, it is obvious that the equation needs to be rescaled so as to absorb the coefficient P. Hence, for that it is assumed that the variables x and y are independent functions of only $\widetilde{x}$ and $\widetilde{y}$, respectively:
$$ \widetilde{x} \propto x$$
$$ \widetilde{y} \propto y$$
<br><br>
Based on this assumption <sup>1</sup>, generalized transformation equations are derived, in other words, the first and second order differentials are defined using the change of variables method:
$$ \frac {\partial \rho}{\partial \widetilde{x}} = \frac {\partial \rho}{\partial x} \frac {\partial x}{\partial \widetilde{x}}  \hspace{2cm}   \frac {\partial \rho}{\partial \widetilde{y}} = \frac {\partial \rho}{\partial y} \frac {\partial y}{\partial \widetilde{y}}$$
<br><br>
$$\frac {\partial^2 \rho}{\partial \widetilde{x}^2} = \left( \frac {\partial x}{\partial \widetilde{x}} \right)^2 \frac {\partial^2 \rho}{\partial x^2} + \left( \frac {\partial y}{\partial \widetilde{x}} \right)^2 \frac {\partial^2 \rho}{\partial y^2} \hspace{2cm}   \frac {\partial^2 \rho}{\partial \widetilde{y}^2} = \left( \frac {\partial y}{\partial \widetilde{y}} \right)^2 \frac {\partial^2 \rho}{\partial y^2} + \left( \frac {\partial x}{\partial \widetilde{y}} \right)^2 \frac {\partial^2 \rho}{\partial x^2}$$ 

<br><br>
Adding the second order differentials and equating expressions to zero results in:

$$\frac {\partial^2 \rho}{\partial \widetilde{x}^2} + \frac {\partial^2 \rho}{\partial \widetilde{y}^2} = \left[ \left( \frac {\partial x}{\partial \widetilde{x}} \right)^2 + \left( \frac {\partial x}{\partial \widetilde{y}} \right)^2 \right] \frac {\partial^2 \rho}{\partial x^2} + \left[ \left( \frac {\partial y}{\partial \widetilde{x}} \right)^2 + \left( \frac {\partial y}{\partial \widetilde{y}} \right)^2 \right] \frac {\partial^2 \rho}{\partial y^2} = 0 $$

<br><br>

Equating the right side of the expression with the equation (2) gives the following relations:
$$ \left( \frac {\partial x}{\partial \widetilde{x}} \right)^2 + \left( \frac {\partial x}{\partial \widetilde{y}} \right)^2 = 1 \hspace{2cm} \left( \frac {\partial y}{\partial \widetilde{x}} \right)^2 + \left( \frac {\partial y}{\partial \widetilde{y}} \right)^2 = P$$

<br><br>

Now, following ansatz can be made based on the previous assumption that the x and y are independent functions of only $\widetilde{x}$ and $\widetilde{y}$, respectively:
$$ x = a\widetilde{x} \hspace{2cm} y = b\widetilde{y}$$

<br><br>

Where a and b are determined by substituing these ansatz in the relations derived before:
$$ \left( \frac {\partial (a\widetilde{x})}{\partial \widetilde{x}} \right)^2 + \left( \frac {\partial (a\widetilde{x})}{\partial \widetilde{y}} \right)^2 = 1 \hspace{2cm} \left( \frac {\partial (b\widetilde{y})}{\partial \widetilde{x}} \right)^2 + \left( \frac {\partial (b\widetilde{y})}{\partial \widetilde{y}} \right)^2 = P$$

$$a^2 = 1 \hspace{2cm} b^2 = P$$

<br><br>
By taking just positive values for convenience the transformations that is needed to be done to turn the equation to Laplace equation are determined:
$$\widetilde{x} = x \hspace{2cm} \widetilde{y} = \frac {y}{\sqrt{P}}$$

$$\partial^2 _ {\widetilde{x}\widetilde{x}} \rho \left( \widetilde{x},\widetilde{y} \right) + \partial^2 _ {\widetilde{y}\widetilde{y}} \rho \left( \widetilde{x},\widetilde{y} \right) = 0 \tag{3}$$

<sup>1</sup> Note: it is possible to derive more general transformation relation assuming that the variables x and y are dependent functions of both $\widetilde{x}$ and $\widetilde{y}$. However it is just too tedious method for simple transformation. 

## Task 2.1
To discretize equation (2), the concentration function $\rho(x,y)$ is defined on a uniform grid, where the uniform grid spacing in the $x-$ and $y-$directions is given by:

$$ \Delta {x} = x_{i+1} - x_{i}$$
$$\Delta {y} = y_{i+1} - y_{i}$$

So, To approximate spatial derivatives, the Taylor expansion of $\rho(x,y)$ is applied in both directions around the point $(x_i, y_j)$:
$$ In\space the \space x \space direction \space \rightarrow  \rho_{i \pm 1,j} = \rho_{i,j} \pm h_{x} \rho^{(1)}_{i,j} + \frac {h^2_{x}}{2!} \rho^{(2)}_{i,j} \pm \frac {h^3_{x}}{3!} \rho^{(3)}_{i,j} + ...$$
<br><br>
$$ In \space the \space y \space direction \space \rightarrow    \rho_{i,j \pm 1} = \rho_{i,j} \pm h_{y} \rho^{(1)}_{i.j} + \frac {h^2_{y}}{2!} \rho^{(2)}_{i,j} \pm \frac {h^3_{y}}{3!} \rho^{(3)}_{i,j} + ...$$

<br><br>

Rearranging these expansions and retaining terms up to second order, the second-order accurate central difference approximations for the second derivatives in space are determined:

$$ \partial^2 _ {xx} \rho _ {i,j} = \frac { \rho_{i + 1, j} - 2\rho_{i,j} + \rho_{i-1,j}} { h_x ^ 2} + \mathcal {O} (h^2_{x})$$
<br><br>
$$ \partial^2 _ {yy} \rho _ {i,j} = \frac { \rho_{i, j + 1} - 2\rho_{i,j} + \rho_{i,j - 1}} { h_y ^ 2} + \mathcal {O} (h^2_{y})$$

<br><br>

Finally, substituting these finite difference expressions into equation (2), the discretized form of the steady-state anisotropic diffusion equation is obtained:
$$ \frac { \rho_{i + 1, j} - 2\rho_{i,j} + \rho_{i-1,j}} { h_x ^ 2} + P \cdot \frac { \rho_{i, j + 1} - 2\rho_{i,j} + \rho_{i,j - 1}} { h_y ^ 2} = \mathcal {O} (h^2_{x}, h^2_{y}) $$

<br><br>

Neglecting the higher-order terms $ \mathcal {O} (h^2_{x}, h^2_{y}) \approx 0 $ , we arrive at the final discretized equation:

$$
\frac{\rho_{i+1,j} - 2\rho_{i,j} + \rho_{i-1,j}}{h_x^2} + P \cdot \frac{\rho_{i,j+1} - 2\rho_{i,j} + \rho_{i,j-1}}{h_y^2} = 0
$$

\frac {P \bar \rho_{F}}{h^2_{y}} \\
\frac {P \bar \rho_{G}}{h^2_{y}} \\
\frac {P \bar \rho_{H}}{h^2_{y}} \\
\frac {P \bar \rho_{I}}{h^2_{y}} \\
\frac {P \bar \rho_{J}}{h^2_{y}}
\end{bmatrix}
$$


The non-zero elements of $ \bar j $ correspond to the left and right Neumann boundary conditions, and each boundary condition appears periodically every 5 entries, corresponding to the vertical columns of the grid. The non-zero elements of $ \bar \rho $ correspond to the upper and lower Dirichlet boundary conditions, while the entries associated with the middle grid points (nodes 6 to 10), which are not adjacent to these boundaries, remain zero.

## Task 4.2

The coefficient matrix obtained is highly sparse and exhibits a clear block-tridiagonal structure when the grid nodes are ordered row-wise. Each block corresponds to a row in the discretized domain, and within each block, the nonzero entries are concentrated along the main and first off-diagonals. This pattern reflects the local coupling inherent in the five-point stencil, where each node interacts only with its immediate neighbors. In the interior of the domain, the matrix is symmetric and diagonally dominant. The matrix’s structure is primarily influenced by the grid spacings $ h_{x} $ and $ h_{y} $ as well as the anisotropy parameter $ P $, which scales the contributions from the y-direction. Specifically, smaller grid spacings result in larger coefficient magnitudes (due to the $ \frac {1}{h^2_{x}} $ and $ \frac {1}{h^2_{y}} $ factors), and variations in $ P $ alter the relative weighting between the x and y diffusion processes.

### Test Problem 4

Before proceeding to compare the numerical and analytical solutions, the analytical solution for the simplified Dirichlet condition will be derived for the stationary diffusion equation in the form of steady Laplace problem:

<br><br>

So, the steady two-dimensional Laplace problem is considered in the unit square is:

<br><br>

$$
\frac{\partial^2 \rho}{\partial x^2} + \frac{\partial^2 \rho}{\partial y^2} = 0, \quad (x, y) \in (0,1) \times (0,1)
$$

<br><br>

which is subjected to the following mixed boundary conditions:

<br><br>

$$
\frac{\partial \rho}{\partial x} \bigg|_{x = 0} = 0, \frac{\partial \rho}{\partial x} \bigg|_{x = 1} = 0, \quad \text{(Neumann, zero flux)}
$$

$$
\rho(x,0) = G_0(x), \quad \rho(x,1) = G_1(x), \quad \text{(Dirichlet)}
$$

<br><br>

Because the Neumann boundary conditions in the $x$-direction impose zero normal derivatives, separation of variables is employed with cosine modes in $x$. The solution is sought in the form:

<br><br>

$$
\rho(x,y) = X(x) Y(y)
$$

<br><br>

Substituting into Laplace's equation gives:

<br><br>

$$
\frac{X''}{X} = - \frac{Y''}{Y} = -\lambda
$$

<br><br>

From the Neumann conditions $X'(0) = X'(1) = 0$, the eigenpairs (the cosine eigenfunctions and associated eigenvalues) are obtained as:

$$
X_n(x) = \cos(n \pi x), \quad \lambda_n = (n\pi)^2, \quad n = 0, 1, 2, \ldots
$$

Solutions in the $y$-Direction:

For the zeroth mode $n = 0$, the equation $ Y_0'' = 0 $ is obtained, which yields the affine solution:

$$
Y_0(y) = a_0 + b_0 y
$$

For the higher modes $n \geq 1$, the ordinary differential equation is obtained:

$$
Y_n'' - (n\pi)^2 Y_n = 0
$$


And its general solution can be written in hyperbolic form:

$$
Y_n(y) = A_n \sinh(n \pi y) + B_n \cosh(n \pi y)
$$

The general solution for the problem is therefore expressed as:

<br><br>

$$
\rho(x,y) = a_0 + b_0 y + \sum_{n=1}^\infty \left[ A_n \sinh(n\pi y) + B_n \cosh(n\pi y) \right] \cos(n \pi x)
$$

<br><br>

The Dirichlet conditions at $y=0$ and $y=1$ are expanded in cosine series (consistent with the Neumann $x$-BCs):

<br><br>

$$
G_0(x) = g_0 + \sum_{n=1}^\infty g_n \cos(n\pi x), \quad G_1(x) = h_0 + \sum_{n=1}^\infty h_n \cos(n\pi x)
$$

<br><br>

where the cosine coefficients are given by:

<br><br>

$$
g_0 = \int_0^1 G_0(x) \, dx, \quad g_n = 2 \int_0^1 G_0(x) \cos(n \pi x) \, dx, \quad n \geq 1
$$

$$
h_0 = \int_0^1 G_1(x) \, dx, \quad h_n = 2 \int_0^1 G_1(x) \cos(n \pi x) \, dx, \quad n \geq 1
$$

<br><br>

The boundary condition at $y=0$, $\rho(x,0) = G_0(x)$, yields:

<br><br>

$$
a_0 = g_0, \quad B_n = g_n, \quad n \geq 1
$$

<br><br>

The boundary condition at $y=1$, $\rho(x,1) = G_1(x)$, leads to two sets of relations. The constant (mean) part gives:

<br><br>

$$
a_0 + b_0 = h_0 \quad \Rightarrow \quad b_0 = h_0 - g_0,
$$

<br><br>

while each cosine mode $n \geq 1$ yields

<br><br>
$$
A_n \sinh(n\pi) + g_n \cosh(n\pi) = h_n, \quad n \geq 1, \quad \Rightarrow \quad A_n = \frac{h_n - g_n \cosh(n\pi)}{\sinh(n\pi)}, \quad n \geq 1
$$

<br><br>

By substitution of the coefficients back into the series, the final analytical solution for the mixed Neumann–Dirichlet boundary condition problem is:

<br><br>

---

$$
\rho(x,y) = g_0 + (h_0 - g_0) y + \sum_{n=1}^\infty \left[ g_n \cosh(n\pi y) + \frac{h_n - g_n \cosh(n\pi)}{\sinh(n\pi)} \sinh(n\pi y) \right] \cos(n \pi x)
$$

---

<br><br>

If the Dirichlet boundaries are constant, $G_0(x) = \rho_\text{bottom}$ and $G_1(x) = \rho_\text{top}$, then all higher cosine coefficients vanish:

<br><br>

$$
g_n = h_n = 0, \quad \forall n \geq 1
$$

<br><br>

The solution reduces to the $n=0$ mode which is a linear profile in $y$, independent of $x$:

<br><br>

$$
\rho(x,y) = \rho_\text{bottom} + (\rho_\text{top} - \rho_\text{bottom}) \, y
$$

<br><br>

This exactly matches the numerical solution in the case $j_\text{left} = j_\text{right} = 0$ (zero flux through the vertical boundaries).


```python
# Grid resolution
N_x = 5
N_y = 5
P = 1
l_x = 1
l_y = 1

# Analytical Solution

# Create the (x, y) coordinate grid
x_vals = np.linspace(0, 1, N_x)
y_vals = np.linspace(0, 1, N_y)
X, Y = np.meshgrid(x_vals, y_vals)

# Define boundary values
rho_upper = 100     # Dirichlet top
rho_lower = 0       # Dirichlet bottom
rho_left_flux = 0   # Neumann left (zero flux)
rho_right_flux = 0  # Neumann right (zero flux)

# Special case: constant top/bottom => only n=0 mode remains
def rho_analytical(x, y, n_limit):
    # Zeroth mode term (linear interpolation in y)
    rho_val = rho_lower + (rho_upper - rho_lower) * y
    
    # If there were non-constant top/bottom profiles, we could add modes here:
    for n in range(1, n_limit + 1):
        g_n = 0.0  # cosine coefficient for bottom BC (zero here)
        h_n = 0.0  # cosine coefficient for top BC (zero here)
        if abs(g_n) > 1e-14 or abs(h_n) > 1e-14:
            A_n = (h_n - g_n * np.cosh(n*np.pi)) / np.sinh(n*np.pi)
            rho_val += (g_n * np.cosh(n*np.pi*y) + A_n * np.sinh(n*np.pi*y)) * np.cos(n*np.pi*x)
    return rho_val

# Evaluate the analytical solution at each (x, y) point
Z = np.zeros_like(X)
n_limit = 75  # Number of cosine terms (higher = more accurate if needed)
for i in range(X.shape[0]):  # Loop over rows (y-direction)
    for j in range(X.shape[1]):  # Loop over columns (x-direction)
        Z[i, j] = rho_analytical(X[i, j], Y[i, j], n_limit)

# Plot the analytical solution using contour plot
plt.figure(figsize=(6, 5))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='ρ(x, y)')
plt.title('Analytical Solution (Neumann left/right, Dirichlet top/bottom)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# Numerical Solution (Matching BCs)

# Internal node counts
N_x_int = N_x - 2
N_y_int = N_y - 2

alpha = -2 * (1 / l_x**2 + P / l_y**2)
beta = 1 / l_x**2
gamma = P / l_y**2

# Submatrices
O = np.zeros((N_x_int, N_x_int))
I = gamma * np.eye(N_x_int)
B = alpha * np.eye(N_x_int) + beta * np.eye(N_x_int, k=1) + beta * np.eye(N_x_int, k=-1)

# Neumann BC in x-direction: double beta on corners
B[0, 1] = 2 * beta
B[-1, -2] = 2 * beta

# Build coefficient matrix
block_rows = []
for i in range(N_y_int):
    row_blocks = []
    for j in range(N_y_int):
        if i == j:
            row_blocks.append(B)
        elif abs(i - j) == 1:
            row_blocks.append(I)
        else:
            row_blocks.append(O)
    block_rows.append(row_blocks)
C = np.block(block_rows)

# Build RHS vector
top = np.full(N_x_int, rho_upper)
bottom = np.full(N_x_int, rho_lower)
middle = np.zeros(N_x_int * (N_y_int - 2))
rho_raw = np.concatenate([top, middle, bottom])
rho_bar = gamma * rho_raw

j_vec = np.zeros_like(rho_bar)  # No extra term since rho_left, rho_right not fixed

# Solve system
rho_sol = np.linalg.solve(C, j_vec - rho_bar)
rho_middle_2d = rho_sol.reshape(N_y_int, N_x_int)

# Add left/right boundaries from Neumann BC (here simply extend first/last cols)
left_col = rho_middle_2d[:, [0]]
right_col = rho_middle_2d[:, [-1]]
rho_middle_full = np.hstack([left_col, rho_middle_2d, right_col])

# Add top/bottom
rho_full = np.vstack([
    np.full((1, N_x), rho_upper),
    rho_middle_full,
    np.full((1, N_x), rho_lower)
])

# Plot the numerical solution
plt.imshow(rho_full, cmap='viridis', origin='upper')
plt.colorbar(label='ρ(x, y)')
plt.title('Numerical Solution (Neumann left/right, Dirichlet top/bottom)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

When comparing the analytical and numerical solutions for the case with Neumann boundary conditions on the left and right and Dirichlet boundary conditions on the top and bottom, the largest discrepancy is observed across the entire vertical profile when using the coarse 5×5 numerical grid. In the analytical solution, the variation of $\rho$ with y is represented as a smooth linear gradient from $\rho_{bottom} = 0$ to $\rho_{top} = 100$, independent of x. This produces a high-resolution, continuous color transition.

In contrast, the numerical solution on a coarse grid consists of only five discrete horizontal layers, which cannot reproduce the smooth gradient and instead creates a piecewise-constant appearance. This discrepancy is not caused by Fourier truncation errors, as in the previous case, but rather by the low spatial resolution of the finite-difference discretization in the y-direction.

When $ j_{left} = j_{right} = 0 $ and only the upper and lower boundaries have Dirichlet conditions, the solution becomes uniform along the x-direction. This is because there is no flux in the x-direction, and the concentration gradient in x vanishes. Therefore, the problem can be reduced from a 2D system to a 1D system in the y-direction, where a simple linear profile is expected. The finite difference equation for a point $\rho_{k}$ in the y-direction becomes:

- **At the boundaries** (e.g. bottom or top nodes):

$$ \rho_{bottom} - 2\rho_{k} = - \bar \rho_{Dirichlet condition} $$

- **At internal nodes** (central difference in y):

$$ \rho_{top} + \rho_{bottom} -2\rho_{k} = 0$$

The solution is a **linear interpolation** between the two boundary values. Once the 1D solution is computed, it can be replicated across all x-columns to construct the full 2D field. This significantly reduces computation time without affecting the accuracy of the result.

```python
# Define the number of nodes in the x and y directions (grid resolution)
N_y = 100         # Number of nodes in y-direction (vertical)
N_x = 100         # Number of nodes in x-direction (horizontal, for replication)

# Define Dirichlet boundary conditions
rho_top = 100     
rho_bottom = 0    

# Create a 1D linear profile from bottom to top
rho_1d = np.linspace(rho_bottom, rho_top, N_y)  # Linear profile

# Expand into 2D by copying the same column across N_x columns
rho_2d = np.repeat(rho_1d, N_x, axis=0)
rho_full = np.vstack(rho_2d)  # Stack into 2D format for plotting

# Plot the resulting concentration field
plt.figure(figsize=(6, 5))
plt.imshow(rho_full, cmap='viridis', origin='lower', aspect='auto')
plt.colorbar(label='ρ(x, y)')
plt.title('Reduced 1D Solution Replicated in 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Additional Test

To compare the numerical and analytical solutions, the analytical solution for the simplified Dirichlet condition will be derived for the stationary diffusion equation in the form of laplace equation for isotropic case:

<br><br>

$$ \partial^2_{xx} \rho (x,y) + \partial^2_{yy} \rho(x,y)=0 \tag{2}$$

<br><br>

Where the boundary conditions are fixed:

<br><br>

$$ \rho \left(0,y \right) = 0, \quad \rho \left(1,y \right) = 0, \quad \rho \left(x,0 \right) = 0, \quad \rho \left(x,1 \right) = f(x)$$

<br><br>

Since, the diffusion equation derived for the principal axes, separation of variables can be applied to solve the PDE:

<br><br>

$$ \rho (x,y) = X(x) \cdot Y(y)$$

<br><br>

$$ \partial^2_{xx} \rho (x,y) = Y(y) \frac {\partial^2 X (x)} { \partial x^2} \quad \partial^2_{yy} \rho (x,y) = X(x) \frac {\partial^2 Y (y)} { \partial y^2}$$

<br><br>

$$ Y(y) \frac {\partial^2 X (x)} { \partial x^2} + X(x) \frac {\partial^2 Y (y)} { \partial y^2} = 0 $$

<br><br>

$$ \frac{1} {X(x)} \frac {\partial^2 X (x)} { \partial x^2} = - \frac {1} {Y(y)} \frac {\partial^2 Y (y)} { \partial y^2} = - \lambda $$

<br><br>

Since RHS and LHS are the functions of different variables they both needs to equal to constant to be equal to each other, which results in the two simple 2nd Order ODE:

<br><br>

$$ \frac{1} {X(x)} \frac {\partial^2 X (x)} { \partial x^2} = - \lambda \quad \frac {1} {Y(y)} \frac {\partial^2 Y (y)} { \partial y^2} = \lambda $$

<br><br>

By applying ansatz solutions, or Laplace/Fourier transformation method results in:

<br><br>

$$ X(x) = sin(n \pi x) \quad Y(y) = sinh(n \pi y) $$

<br><br>

The general solution is a sum of all possible separated solutions:

<br><br>

$$ \rho (x,y) = \sum_{n=1}^{\infty} { A_n sin(n \pi x) sinh(n \pi y) } $$

<br><br>

Applying the boundary condition $ \rho \left(x,1 \right) = f(x)$:

$$ \rho (x,1) = \sum_{n=1}^{\infty} { A_n sin(n \pi x) sinh(n \pi) } = f(x)$$

<br> <br>

Letting:

$$ B_n = A_n \sinh(n \pi) \quad \Rightarrow \quad f(x) = \sum_{n=1}^{\infty} B_n \sin(n \pi x) $$

Finding the Fourier sine coefficients:

$$ B_n = 2 \int_0^1 f(x) \sin(n \pi x)\, dx $$

Solving for $ A_n $:

$$ A_n = \frac{B_n}{\sinh(n \pi)} = \frac{2}{\sinh(n \pi)} \int_0^1 f(x) \sin(n \pi x)\, dx $$

<br><br>

Finally plugging the $ A_n $ back into the full solution:

$$ \rho(x, y) = \sum_{n=1}^\infty \left[ \frac{2}{\sinh(n \pi)} \int_0^1 f(x) \sin(n \pi x)\, dx \right] \sin(n \pi x) \sinh(n \pi y) $$

<br><br>

Let’s solve it for the special case $f(x) = 100$:

<br><br>

$$ A_n = \frac{2}{\sinh(n \pi)} \int_0^1 100 \sin(n \pi x)\, dx  = \frac{400}{n \pi \sinh(n \pi)} \rightarrow \space n = 1, 3, 5, \cdots$$
$$ A_n = 0 \rightarrow \space n = 2, 4, 6, \cdots$$

<br><br>

---

Final Analytical Solution (for $f(x) = 100$):

<br><br>

$$ \rho (x,y) = \sum_{n = 1, 3, 5, \cdots}^{\infty} { \frac{400}{n \pi \sinh(n \pi)} sin(n \pi x) sinh(n \pi y) } $$

---

```python
N_x = 100
N_y = 100


# Analytical Solution

# Create the (x, y) coordinate grid
x_vals = np.linspace(0, 1, N_x)
y_vals = np.linspace(0, 1, N_y)
X, Y = np.meshgrid(x_vals, y_vals)

# Define the analytical solution function

def rho(x,y, n_limit):
    rho_val = 0
    for n in range(1, n_limit + 1, 2):
        term = ( 400 / ( n * np.pi *np.sinh (n * np.pi))) * np.sin( n*np.pi*x) * np.sinh( n*np.pi*y)
        rho_val += term
    return rho_val


# Evaluate the analytical solution at each (x, y) point in the grid
Z = np.zeros_like(X)
n_limit = 75 # Number of terms in Fourier series (higher = more accurate)

for i in range(X.shape[0]): # Loop over rows (y-direction)
    for j in range(X.shape[1]): # Loop over columns (x-direction)
        Z[i, j] = rho(X[i, j], Y[i, j], n_limit)

# Plot the analytical solution using contour plot
plt.figure(figsize=(6, 5))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='ρ(x, y)')
plt.title(' Analytical Solution of 2D Diffusion Equation ')
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# Numerical Solution (Matching Dirichlet BCs)
O = np.zeros((N_x, N_x))
I = gamma * np.eye(N_x)

# Construct the B matrix for each internal row of the grid
I_1 = np.eye(N_x)
I_1 *= alpha  # Multiply inner diagonal entries (2,2), (3,3), (4,4),... by 5

# Add off-diagonal beta terms to the B matrix for x-direction neighbors

B = I_1 + beta * np.eye(N_x, k=1) + beta * np.eye(N_x, k=-1)

B[0, 1] = 2 * beta
B[-1, -2] = 2 * beta

# Construct the full coefficient matrix C using submatrices
block_rows = []
for i in range(N_y-2):
    row = []
    for j in range(N_y-2):
        if i == j:
            row.append(B)   # Diagonal block
        elif abs(i - j) == 1:
            row.append(I)   # Upper and Lower diagonal blocks for coupling with vertical neighbors
        else:
            row.append(O)
    block_rows.append(row)
C = np.block(block_rows)


# Define Dirichlet boundary values (same as in analytical solution)
rho_upper = 100
rho_lower = 0
rho_left = 0
rho_right = 0

# Number of unknowns (internal nodes only)
N_x_internal = N_x - 2  # exclude left and right boundaries
N_y_internal = N_y - 2  # exclude top and bottom
num_unknowns = N_x_internal * N_y_internal

# Define the submatrices for the full coefficient matrix
O = np.zeros((N_x_internal, N_x_internal))
I = gamma * np.eye(N_x_internal)


I_1 = np.eye(N_x_internal)
I_1 *= alpha

B = I_1 + beta * np.eye(N_x_internal, k=1) + beta * np.eye(N_x_internal, k=-1)

# Full coefficient matrix C
block_rows = []
for i in range(N_y_internal):
    row = []
    for j in range(N_y_internal):
        if i == j:
            row.append(B)
        elif abs(i - j) == 1:
            row.append(I)
        else:
            row.append(O)
    block_rows.append(row)
C = np.block(block_rows)

# Build RHS vector
# Contributions from Dirichlet top and bottom
top = np.full(N_x_internal, rho_upper)
bottom = np.full(N_x_internal, rho_lower)
middle = np.zeros(N_x_internal * (N_y_internal - 2))
rho_raw = np.concatenate([top, middle, bottom])
rho_bar = gamma * rho_raw  # scaled

# Contributions from left and right Dirichlet BCs
j_vec = np.zeros(num_unknowns)

for row in range(N_y_internal):
    idx_start = row * N_x_internal
    j_vec[idx_start]     += beta * rho_left   # left boundary
    j_vec[idx_start + N_x_internal - 1] += beta * rho_right  # right boundary

# Solve system
rho = np.linalg.solve(C, j_vec - rho_bar)
rho_middle = rho.reshape(N_y_internal, N_x_internal)

# Reconstruct full grid including boundaries
left_col = np.full((N_y_internal, 1), rho_left)
right_col = np.full((N_y_internal, 1), rho_right)
rho_middle_full = np.hstack([left_col, rho_middle, right_col])
rho_full = np.vstack([np.full(N_x, rho_upper), rho_middle_full, np.full(N_x, rho_lower)])

# Plot
plt.imshow(rho_full, cmap='viridis', origin='upper')
plt.colorbar(label='ρ(x, y)')
plt.title('Numerical Solution of 2D Diffusion Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
